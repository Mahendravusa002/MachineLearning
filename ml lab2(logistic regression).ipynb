```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as mtp
dataset=pd.read_csv("User_Data.csv")
```

```python
x=dataset.iloc[:,[2,3]].values
y=dataset.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test) 
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression
cla=LogisticRegression(random_state=0)
cla.fit(x_train,y_train)
```

```python
y_pred=cla.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix
```

```python
from matplotlib.colors import ListedColormap
x_set,y_set=x_train,y_train
x1,x2=nm.meshgrid(nm.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                  nm.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
mtp.contourf(x1,x2,cla.predict(nm.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75,cmp=ListedColormap(('purple','green')))
mtp.xlim(x1.min(),x1.max())
mtp.ylim(x2.min(),x2.max())
for i,j in enumerate(nm.unique(y_set)):
    mtp.scatter(x_set[y_set==j,0],x_set[y_set==j,1],c=ListedColormap(('purple','green'))(i),label=j)
mtp.title("Logistic Regression(Traing set)")
mtp.xlabel('Age')
mtp.ylabel('Estimated salary')
mtp.legend()
mtp.show()
```
